```python
import json
import pandas as pd
```

```python
def get_coulmns_names(schemas, ds_name, sorting_key='column_position'):
    column_details = schemas[ds_name]
    columns = sorted(column_details, key=lambda col: col[sorting_key])
    return [col['column_name'] for col in columns]
```

```python
schemas = json.load(open('../data/retail_db/schemas.json'))
```

```python
orders_columns = get_coulmns_names(schemas, 'orders')
```

```python
orders = pd.read_csv(
    '../data/retail_db/orders/part-00000',
    names = orders_columns
)
```

```python
customers_columns = get_coulmns_names(schemas, 'customers')
```

```python
customers = pd.read_csv('../data/retail_db/customers/part-00000', names=customers_columns)
```
